This dataset is the Allstate Purchase Prediction Challenge, currently on Kaggle.  Its observational data of different Allstate customers as they go through the online shopping process, with the last row of data per customer indicating their actual choice versus options they viewed in prior rows.

As a customer shops an insurance policy, he/she will receive a number of quotes with different coverage options before purchasing a plan. This is represented as a series of rows that include a customer ID, information about the customer, information about the quoted policy, and the cost. The objective is to predict the purchased coverage options using a limited subset of the total interaction history. Ultimate objective is to present a customer with most likely purchase options sooner to ensure customer stays on website and purchases rather than going elsewhere..

Kaggle Question:
Using a customer’s shopping history, can you predict what policy they will end up choosing?
OUR MINI OBJECTIVE for this homework is to determine, based on other columns, whether a customer will choose 1 for option B or 0 for option B.  Allstate has NOT chose to tell us what Option B is, but we should be able to use Logistic Regression to determine customer likelihood of choosing this option.

The dataset has about 24 features and 665000 rows.. awesome, the issue here is narrowing the features down to the appropriate subset.
The dataset is appropriate to a binary approach because the response (choose or not) is binary, as well as many other options (whether or not they chose other options, etc). 
I chose this dataset because its ready to analyze, big enough and important enough to be on Kaggle, and maybe I'll win 50 grand.
